# Initialising, optimizing and sampling the two models:

Trying a single planet model for HD114082 - a known (albeit distant) planet-host

```python
mod=AstrometricModel("TIC441546821", omegas_prior="p&m", ecc_prior="sqrtesinomega", planet_mass_prior='log', data_release='DR5') #Trying a high mass and a bright/nearby star, and using the reparameterised eccentricity

mod.star_from_tic()
mod.init_planet('b', mpl_mjup=5.1, empl_mjup=1.3, per_d=481.238, eper_d=120,  t0_jd=2460200, et0_jd=0.04, b=0.4, eb=0.1)
mod.init_gaia_data()

mod.optimize_model(modname='w_planets')
mod.optimize_model(modname='no_planets')
```

```python
#Due to the difficulty of sampling this parameter space, I have boosted this for the with-planet case:
mod.sample_models(modnames=['w_planets'],regularization_steps=12,n_draws=500,n_tune=150,save=False)
mod.sample_models(modnames=['no_planets'],regularization_steps=6,n_draws=800,n_tune=200,save=False)
```

# Initialising, optimizing and sampling the two models:

Trying a 2-planet model for HD114082 - a known (albeit distant) planet-host

```python
mod=AstrometricModel("TIC441546821", omegas_prior="p&m",ecc_prior="kipping") #Trying a high mass and a bright/nearby star, and using the reparameterised eccentricity

mod.star_from_tic()
mod.init_planet('b', mpl_mjup=6, empl_mjup=1.5, per_d=500, eper_d=150,  t0_jd=2460200, et0_jd=0.04, b=0.4, eb=0.1)
mod.init_planet('c',mpl_mjup=11,empl_mjup=2.5,per_d=280,eper_d=80,t0_jd=2460000,et0_jd=0.04,b=0.6,eb=0.1,
                ecc=0.01)
mod.init_gaia_data()

mod.optimize_model(modname='w_planets')
mod.optimize_model(modname='no_planets')
```

```python
#Due to the difficulty of sampling this parameter space, I have boosted this for the with-planet case:
mod.sample_models(modnames=['w_planets'],regularization_steps=12,n_draws=1500,n_tune=500)
mod.sample_models(modnames=['no_planets'],regularization_steps=6,n_draws=800,n_tune=200)
```

# Save location

This is hard-coded to be in a folder called "data" which is in the directory of the `tpgaia.py` file (a new folder is generated for each star name).

Typically things are saved with one of two prefixes. One is model-dependent and includes a number (i.e. changes each time the model gets re-run) which is good as it means we won't overwrite old stuff. The other is model-independent and should include files which are model-independent (e.g. the Gaia GOST data)

```python
import exoplanet as xo
from astropy.time import Time
from astropy.coordinates import Angle
from astropy import constants
from astropy import units as u
import numpy as np
import pandas as pd
import requests
import xml.etree.ElementTree as ET
from htof.sky_path import parallactic_motion, earth_ephemeris, earth_sun_l2_ephemeris
from htof.main import Astrometry

from htof import parse
from scipy import interpolate
from matplotlib import pyplot as plt
import pymc as pm
import pymc_ext as pmx
import arviz as az
#%matplotlib inline
```

```python
#initialize some parameters for the simulated data
input_p  = 542.0
input_mp = 10
input_ecc = 0.1
input_omega = 0.82*np.pi
input_Omega = 1.25127*np.pi #all angles in radians
input_ttrans = 2460000.136
input_b = 0.4
input_Rs = 1.0
input_Ms = 1.0
input_Gmag = 9.0
input_plx = 1000/30 #parallax in marcsec
RA_deg = 37.1234 #mean position of star
DEC_deg = -39.183 
```

```python
#useful conversions

_degtorad = np.pi/180
_radtomas = (180 * 3600 * 1000) / np.pi
_mastorad = np.pi / (180 * 3600 * 1000)
_kmps_to_aupyr = (u.year.to(u.s) * u.km.to(u.m)) / constants.au
_degtomarcsec=3600*1000


#get predicted sky coordinate positions for observation of the star along with times, scan angle and parallaxes
url = f"https://gaia.esac.esa.int/gost/GostServlet?ra="+str(RA_deg)+"&dec="+str(DEC_deg)+"&service=1"
with requests.Session() as s:
    s.get(url)
    headers = {"Cookie": f"JSESSIONID={s.cookies.get_dict()['JSESSIONID']}"}
    response = s.get(url, headers=headers, timeout=180)
    dat = response.text

root = ET.fromstring(dat)
columns = ["Target", "ra[rad]", "dec[rad]", "ra[h:m:s]", "dec[d:m:s]", "ObservationTimeAtGaia[UTC]",
            "CcdRow[1-7]", "zetaFieldAngle[rad]", "scanAngle[rad]", "Fov[FovP=preceding/FovF=following]",
            "parallaxFactorAlongScan", "parallaxFactorAcrossScan", "ObservationTimeAtBarycentre[BarycentricJulianDateInTCB]"]
rows = []
name = root.find('./targets/target/name').text #extract data
raR = root.find('./targets/target/coords/ra').text
decR = root.find('./targets/target/coords/dec').text
raH = root.find('./targets/target/coords/raHms').text
decH = root.find('./targets/target/coords/decDms').text
for event in root.findall('./targets/target/events/event'):
    details = event.find('details')
    observationTimeAtGaia = event.find('eventUtcDate').text
    ccdRow = details.find('ccdRow').text
    zetaFieldAngle = details.find('zetaFieldAngle').text
    scanAngle = details.find('scanAngle').text
    fov = details.find('fov').text
    parallaxFactorAl = details.find('parallaxFactorAl').text
    parallaxFactorAc = details.find('parallaxFactorAc').text
    observationTimeAtBarycentre = event.find('eventTcbBarycentricJulianDateAtBarycentre').text
    rows.append([name, raR, decR, raH, decH, observationTimeAtGaia, ccdRow,
                  zetaFieldAngle, scanAngle, fov, parallaxFactorAl, parallaxFactorAc, observationTimeAtBarycentre])
gaia_obs_data = pd.DataFrame(rows, columns=columns)

gaia_ALplxfact=gaia_obs_data["parallaxFactorAlongScan"]
```

```python
def get_Gaia_error(GAIAmag):
    """
    Using the interpolation function to estimate along scan RMS error.
    """
    ast_std=pd.read_csv("Gaia_AlongScan_Uncertainties.csv",header=None)
    interp=interpolate.interp1d(ast_std[0].values,ast_std[1].values)
    return interp(GAIAmag)
    
# The Gaia data is now loaded in as "gaia_obs_data" - accessing most important columns:
gaia_t=Time(gaia_obs_data["ObservationTimeAtBarycentre[BarycentricJulianDateInTCB]"].values.astype(float),format='jd')
gaia_scanAng_rad=gaia_obs_data["scanAngle[rad]"].values
gaia_ALplxfact=gaia_obs_data["parallaxFactorAlongScan"].values.astype(float)
gaia_ACplxfact=gaia_obs_data["parallaxFactorAcrossScan"].values.astype(float)


# Normal triad, defined at the reference epoch., gives the cartesian base vectors for a certain reference frame
p = np.array([-np.sin(RA_deg*_degtorad), np.cos(RA_deg*_degtorad), 0.0])
q = np.array([-np.sin(DEC_deg*_degtorad) * np.cos(RA_deg*_degtorad), -np.sin(DEC_deg*_degtorad) * np.sin(RA_deg*_degtorad), np.cos(DEC_deg*_degtorad)])
r = np.array([np.cos(DEC_deg*_degtorad) * np.cos(RA_deg*_degtorad), np.cos(DEC_deg*_degtorad) * np.sin(RA_deg*_degtorad), np.sin(DEC_deg*_degtorad)])

bO_bcrs = earth_sun_l2_ephemeris(gaia_t.jyear) #movement of Lagrange point?

# Calculate the Roemer delay, take units into account.
tB = gaia_t + np.dot(r, bO_bcrs) * constants.au / constants.c / u.year.to(u.s)
uO_init = np.repeat(r, gaia_t.size).reshape((r.size, gaia_t.size))

gaia_tfine = Time(np.arange(np.min(gaia_t).jd,np.max(gaia_t).jd,1),format='jd')
bO_bcrs_fine=earth_sun_l2_ephemeris(gaia_tfine.jyear)
tB_fine = gaia_tfine + np.dot(r, bO_bcrs_fine) * constants.au / constants.c / u.year.to(u.s)
uO_init_fine = np.repeat(r, gaia_tfine.size).reshape((r.size, gaia_tfine.size))

#Also initialising the expected Gaia error:
gaia_error = get_Gaia_error(input_Gmag)
```

```python
def astrometric_signal(m_planet, m_star,semia,dist): #return the astrometric signal, i.e angle we observe in arcsec
        """
        beware: m_planet and M-star must have same units, semia in AU, dist in pc
        """
        return m_planet/m_star *semia/dist

def SNR(alpha,sigma): #sigma from gaia_error
    return alpha/sigma

def semimajax(period): #return the semimajor axis in AU assuming mstar>>mplanet
    return (7.496e-6*period**2)**(1/3)

def planetyn(trace): #calculate 3-sigma, if below 13Mjup then presence of a planet can be inferred
    top=np.percentile(trace.posterior["mpl"],99.7)
    yn=bool(13>top)
    return top,yn
```

```python
_marcsectodeg=1/(1000*3600)
"""
initialize proper motion for simulated system
"""
init_pm_ra=0.05 #marcsec per year
init_pm_dec=0.05 #marcsec per year

#calculates the RA & DEC due to proper motion of the system
rat=RA_deg + init_pm_ra*_marcsectodeg*(gaia_t.jd-Time(2015.5,format='jyear').jd) 
dect=DEC_deg + init_pm_dec*_marcsectodeg*(gaia_t.jd-Time(2015.5,format='jyear').jd)


```

```python

"""
function to generate simulated alongscan deviations from predicted positions
"""
def generate_model_data(input_p,input_ttrans,input_b,input_ecc,input_Rs,input_Ms,input_mp,input_omega,input_Omega,input_plx,gaia_scanAng_rad):
    orbit = xo.orbits.KeplerianOrbit( #initialize the artificial model
        period=input_p, #period
        t0=input_ttrans, #time of transit
        b=input_b, #impact parameter
        ecc=input_ecc, #eccentricity
        rstar=input_Rs, #radius of star
        mstar=input_Ms, #star mass
        m_planet=input_mp, #planet mass
        m_planet_units=u.Mjup, #specify which unit we are taking, here jupiter masses
        omega=input_omega, #argument of perihel passage
        Omega=input_Omega) #argument of ascending node

    t0_ra=orbit.get_star_position(Time(2015.5,format='jyear').jd,parallax=input_plx)[0].eval() #get star position at a certain time  which will be our reference
    t0_dec=orbit.get_star_position(Time(2015.5,format='jyear').jd,parallax=input_plx)[1].eval()
    
    true_ra=orbit.get_star_position(gaia_t.jd,parallax=input_plx)[0].eval()
    true_dec=orbit.get_star_position(gaia_t.jd,parallax=input_plx)[1].eval()
    true_dist=orbit.get_star_position(gaia_t.jd,parallax=input_plx)[2].eval()
    
    # print(orbit.get_star_position(gaia_t.jd)[0].eval())

    # print(orbit.get_star_position(gaia_t.jd)[0].eval())
    # plt.scatter(true_ra-t0_ra,true_dec-t0_dec,c=gaia_t.jyear,s=2)
    # #plt.scatter(true_ra,true_dec,c=gaia_t.jyear,s=2)
    # plt.colorbar()
    # plt.title("Offset from mean position assuming 1 planet")
    # plt.show()
    #y = Deltadec_pl*tt.sin(self.gaia_scanAng_rad)+(model_params['DEC_offset_mas']+model_params['Deltadec_par_pm']+tt.sum(tt.stack([model_params['Deltara_pl'][pl] for pl in self.planets],axis=0),axis=0))*tt.cos(self.gaia_scanAng_rad)
    # plt.subplot(211)
    
    alongscan_y = (true_ra-t0_ra)*np.sin(gaia_scanAng_rad.astype(float))+(true_dec-t0_dec)*np.cos(gaia_scanAng_rad.astype(float)) #rotation/projection onto AL
    # plt.title("AL-scan measurement in rad")
    # plt.scatter(gaia_scanAng_rad.astype(float),alongscan_y,c=true_dec)
    # plt.xlabel("scanangle")
    # plt.ylabel("AL measurement")
    # plt.colorbar(label='dec')
    # plt.subplot(212)
    # plt.scatter(gaia_scanAng_rad.astype(float),alongscan_y,c=true_ra)
    # plt.xlabel("scanangle")
    # plt.ylabel("AL measurement")
    # plt.colorbar(label='ra')
    
    # plt.show()
    # alongscan_y_werr = alongscan_y + np.random.normal(np.zeros(len(alongscan_y)),np.tile(gaia_error,len(alongscan_y)))
    # plt.plot(alongscan_y,'.', label="AL without error")
    # plt.plot(alongscan_y_werr,'.',alpha=0.3,label="AL with error")
    # plt.title("AL data with and without normally distributed errors")
    # plt.legend()
    # plt.xlabel("array index")
    # plt.ylabel("AL")
    # plt.show()

    return alongscan_y

```

Fitting model initialisation:

```python
def model_fit(input_p,input_ttrans,input_plx,gaia_t,gaia_scanAng_rad,alongscan_y,gaia_error):
    with pm.Model():
    
        #initialize priors
        # Period, semi-major axis, eccentricity, and t0
        period = pm.Normal("period", mu=np.random.normal(input_p,0.01), sigma=0.02)
        ecc = xo.distributions.eccentricity.kipping13("ecc",fixed=True)
        t0 = pm.Normal("tperi", mu=np.random.normal(input_ttrans,0.01), sigma=0.01)
        
        # For astrometric orbits, a good choice of parameterization can be
        # (Omega ± omega) / 2
        plus = pmx.angle("plus")
        minus = pmx.angle("minus")
        Omega = pm.Deterministic("Omega", plus + minus)
        omega = pm.Deterministic("omega", plus - minus)
        
    
        # We'll use a uniform prior on cos(incl)
        b = pm.Uniform("cos_incl", lower=0, upper=1.0)
        mstar=pm.Normal("Mstar",mu=1.0,sigma=0.1) #assume Sun-like star
        rstar=pm.Normal("Rstar",mu=1.0,sigma=0.01)
        #incl=pm.Normal("incl",mu=np.pi/2,sigma=0.1)
    
        logm_planet = pm.Uniform("logmpl", lower=np.log(0.25), upper=np.log(150)) #uniform planet mass distr in log space
        m_planet = pm.Deterministic("mpl", pm.math.exp(logm_planet))
        
    
        ra  = pm.Normal("ra",mu=RA_deg,sigma=0.0001)
        dec = pm.Normal("dec",mu=DEC_deg,sigma=0.0001)
        # Then we define the orbit
        orbit = xo.orbits.KeplerianOrbit(
            t0=t0,
            period=period,
            b=b,
            mstar=mstar,
            rstar=rstar,
            ecc=ecc,
            omega=omega,
            Omega=Omega
        )

        ###TO DO: we have weird variables occuring (_minus_angle1 etc)(appear to just be there for some reason(omega?)-> related to the 
                        #angles generated by pmx.angle, see example below
        #         the fitted mass is way off, although we might as well only take a certain percentile
        #    Have wierd behavior of AL_offset as all values are zero-> try with similar setting in other loop (example below to try and reproduce!
        #  ->appears also in other woking module
        # therefore not considered influential on result
        
    
        #print(orbit.get_planet_position(Time(2015.5,format='jyear').jd,parallax=input_plx))
        ra_at_t0=pm.Deterministic("ra_at_t0", orbit.get_star_position(Time(2015.5,format='jyear').jd, parallax=input_plx)[0]) #position of star at given time, define as zero-point
        dec_at_t0=pm.Deterministic("dec_at_t0", orbit.get_star_position(Time(2015.5,format='jyear').jd, parallax=input_plx)[1]) #w.r.t which movement is measured
        #print(ra_at_t0.eval())
        ra_offset_from_bary = pm.Deterministic("ra_offset_from_bary", orbit.get_star_position(gaia_t.jd,parallax=input_plx)[0]) #position of star as fct of time
        dec_offset_from_bary = pm.Deterministic("dec_offset_from_bary", orbit.get_star_position(gaia_t.jd,parallax=input_plx)[1])
    
        ra_offset_from_t0 = pm.Deterministic("ra_offset_from_t0", ra_offset_from_bary-ra_at_t0) #offset from position of star at t0
        dec_offset_from_t0 = pm.Deterministic("dec_offset_from_t0", dec_offset_from_bary-ra_at_t0)
    
        AL_offset = pm.Deterministic("AL_offset", (ra_offset_from_t0)*pm.math.sin(gaia_scanAng_rad.astype(float))+(dec_offset_from_t0)*pm.math.cos(gaia_scanAng_rad.astype(float)))
        #gives AL amplitude as fct of time
    
        
        pm.Normal(                    # observed angle movement in the sky,likelihood
            "theta_obs",
            mu=AL_offset,
            sigma=gaia_error,
            observed=alongscan_y)
        
        init_soln = pm.find_MAP()
        # #print(init_soln)
        trace = pm.sample(
            tune=1000,
            draws=1000,
            cores=2,
            chains=2,
            target_accept=0.95,
            init="adapt_full",
            initvals=init_soln
        )
        return trace
```
